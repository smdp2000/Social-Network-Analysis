Random Network

```r
library(igraph)
library(ggplot2)
```

```r
make_random = function(num,prob)
{
    G = make_empty_graph(n=num)
    for (i in 1:(num - 1)){
        for(j in (i+1):num){
            r = runif(1)
            if(r<=prob){
                G = G + edge(i,j)
            }
        }
    }
    return(G)
}
```

```r
G1 = make_random(100, 0.25)
d1 = degree(G1)
plot1 = table(d1) 
plot(plot1 / sum(plot1), xlim=c(15,36), ylim=c(0, 0.15), xlab="Degree", ylab="Frequency")
print(degree_distribution(G1))
```
